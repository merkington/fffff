```python
import requests
import json
import numpy as np
from datetime import datetime

date = datetime(2007, 1, 1)

def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)
    
parameters = {
    "lat": 52.629729,
    "lng": -1.131592  
}

response = requests.get("https://data.police.uk/api/stops-street" ,params=parameters)
jdata = response.json()
jprint(response.json())

ctype = np.array([[0,1,2,3],[0,0,0,0]]) # crime type

for i in range(len(jdata)):
  ch = (jdata[i]['age_range'])#counter
  if ch == "10-17": ctype[1,0] += 1
  if ch == "18-24": ctype[1,1] += 1
  if ch == "25-34": ctype[1,2] += 1
  if ch == "over 34": ctype[1,3] += 1

#  print(ch)
print(ctype)

print(response.status_code)
```
